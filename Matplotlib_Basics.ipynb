```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpt = np.array([0,12])
ypt = np.array([0,200])
plt.plot(xpt,ypt,marker = 'o')  # marker darkens the points
```

```python
# Now if you wanna plot without line then just add 'o'
xpt = np.array([0,12])
ypt = np.array([0,200])
plt.plot(xpt,ypt,'o')
```

```python
x = np.array([34,55,44,67,23,67,89,90,12,95])
y = np.array([66,77,88,99,45,3,23,67,49,82])
x = np.sort(x)
y = np.sort(y)
plt.plot(x,y, marker = '*')
plt.title('Student Stats')
plt.xlabel('Students')
plt.ylabel('Marks in Maths')
plt.grid()
plt.show()
```

```python
x = np.array([10,20,30,40,50,60,75,90,95,100])
y = np.array([12,24,36,44,60,72,84,96,108,120])

plt.plot(x,y)
font1 = { 'family': 'sans', 'color':'blue','size':15}
font2 = {'family': 'serif', 'color':'green','size':25}
plt.title('Finding lost planets' , loc = 'left' , fontdict = font2)
plt.xlabel('Exoplanets discovered', fontdict = font1)
plt.ylabel('Light Intensity', fontdict = font1 )
plt.grid(color = 'red' , linestyle = '--' )
```

```python
x = np.array([34,55,44,67,23,67,89,90,12,95])
y = np.array([66,77,88,99,45,3,23,67,49,82])
x = np.sort(x)
y = np.sort(y)
plt.plot(x,y, 'o:c')  # c is cyan color
# Here o:g is a string format tool i.e. fmt whose syntax says marker|line|color
```

```python
# Adding bars
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 10, 2, 7])

plt.bar(x,y , color = 'hotpink')
```

```python
# Adding pie
pts = np.array([35, 25, 25, 15])
dept = ["Marketing", "Web-Dev", "App-Dev", "Creatives"]

plt.pie(pts , labels = dept)
```
