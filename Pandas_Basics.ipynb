```python
newdf2 = newdf
# Here newdf2 is a view of newdf i.e. when values of newdf2 changes newdf also changes
# This is because newdf2 is pointing to newdf
newdf2[0][0] = 555
```

```python
newdf.head()
# here the first element has changed to 555
```

```python
newdf3 = newdf.copy()
# We made newdf3 a copy of newdf
# This means that if you change values of newdf3 , this won't change content in newdf
newdf3[0][0] = 99
print(newdf)      # 1st element doesn't change
```

```python
newdf.columns = list('ABCDE')
# You can change values with loc function
newdf.loc[0][0] = 234
newdf.head(2)
```

```python
# loc method
# You can locate rows by their index number.
print(newdf.loc[1])

# You can also locate multiple rows at once by providing a list of index numbers
print(newdf.loc[[2,5,8]])

# You can locate specific cells by specifying both the row and column
print('Required element = ',newdf.loc[1]['A'])
```

```python
newdf.loc[0,'cancel'] = 37
print(newdf.head())

# Since , I created cancel column unnecessarily , I can delete it by drop method :
# drop method will return a new df and doesn't change the original one , since it returns a copy
# so save it in newdf only
newdf = newdf.drop('cancel',axis=1)   # Here by default , axis is
newdf
```

```python
newdf.loc[[2,4],:]
# Here : means all the rows
```

```python
# You can also add conditions in loc like
newdf.loc[(newdf['B'] < 0.02) & (newdf['C'] > 0.01)]
```

```python
# If you wanna access elements by indexing only then use iloc method
# Remember that loc() is for accessing through row/col names and not index
newdf.iloc[222][2]
```

```python
newdf.iloc[[151,0], [0,1]]
#          [rows],[columns]
```

```python
newdf.head()
newdf.drop(['A','E'], axis=1 inplace=True)
# We know that the above one is a copy , but inplace attribute deletes from the original dataframe
# Also remember that , running it more than once could cause an error
```

```python
newdf.drop([1,3,4],inplace = True , axis = 0)
```

```python
newdf.head(5)
```

```python
# Resetting index since we deleted rows 1,3,4
# Remember that content is deleted permanently , only reserialing is done
newdf.reset_index(drop=True , inplace = True)
# Here drop ensures that another index col doesn't come
newdf
```

```python
# Always use loc or iloc for setting values
newdf.loc[: ,['B']] = None
newdf
```

```python
# dropna() method
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
df
```

```python
# Now , I don't want NaN elements , so :
df = df.dropna()
df
# also attribute like (how='all',axis =1 ) will remove nan only if all the elements of the row contains it
```
